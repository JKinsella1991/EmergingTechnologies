# About Iris Dataset

This dataset is used in the field of machine learning and computer visualization, and is used to classify different species and sub-species of flowers. It does this by visually comparing various parts of the flower, such as the sepal and petal, against it's own knowledgebase. Additionally, we'll also be using the Pandas python data analysis library

## Imports

```python
# Import the pandas package
import pandas as pd

# Load the dataset, we'll use an external source
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")

# Display the data
df
```

```python
# Check what types of data we're outputting
df.dtypes
```

```python
# Next we'll format the data with some headings
df[['sepal_length', 'sepal_width']]
```

```python
# With the same syntax as we've seen previously ([xx:yy]) we can display specific ranges
df[['sepal_length', 'sepal_width']][1:40]
```

## Visualizing the Iris Dataset

In this next section we'll graph and show the difference between each setosa, Virginica, et al. For this we'll again be using MatPlotLib, and a data visualization library for Python called Seaborn.

## More imports

```python
# Import MatPlotLib and Seaborn, each used to plot data

import matplotlib.pyplot as plt

import seaborn as sb

# We'll use this symbol '^' to show the points as we graph
graph_plot = sb.pairplot(df, hue='species', markers='^')
```

As we didn't put any constraints on the output of the graphing function, we get this "explosion" of information, containing overlapping histograms and scatter diagrams.
